```python
from google.cloud import bigquery

# Construct a BigQuery client object.
client = bigquery.Client()

# Define your project ID (replace with your actual project ID)
project_id = "dosewise" 

# Define your SQL query
query = """
    SELECT
        *
    FROM
        `dosewisedb.clinic_table`
    LIMIT 10
"""

# Execute the query
query_job = client.query(query, project=project_id)  # API request - fetches results

# Convert the results to a Pandas DataFrame for easier manipulation
df = query_job.to_dataframe()

# Display the data
print(df.head())
```
